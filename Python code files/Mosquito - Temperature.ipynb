```python
def batchtemp(attr,lim1,lim2):
    from datetime import date, timedelta
    fig2 = figure(figsize=(20,25))
    batchaug=[]
#pdf1 = matplotlib.backends.backend_pdf.PdfPages("/data/dhence/a/pborse2/inputdata/Mosquito data/Batches&Rainfall.pdf")
    for i in range(len(csvlist)):
        plt.subplot(8, 3, i+1)
        if(mosyr.count(csvlist[i].index[1].year)>0):
            y=csvlist[i].index[1].year
            batch2 = batchdf[batchdf['date'].dt.year==y]
            plt.plot(batch2['date'],batch2['batches'], marker='o')
            plt.ylim(0,17)
            plt.ylabel('mosq.')
            ax2 = csvlist[i][attr][batch2['date'].iloc[0]-timedelta(15):batch2['date'].iloc[-1]+timedelta(15)].plot(secondary_y=True, color='g')
            print(csvlist[i][attr].max())
            ax2.set_ylabel(attr)
            ax2.set_ylim(lim1,lim2)
            plt.legend('temp')
            batchaug.append((batch2[batch2['date'].dt.month==8]['batches'].sum()/batch2['batches'].sum())*100)
    return batchaug
```

```python
from datetime import timedelta
lagdays=[]
for j in range(28):
    laglist=[]
    for i in range(len(csvlist)):
        if(mosyr.count(csvlist[i].index[1].year)>0):
            y=csvlist[i].index[1].year
            batch3 = batchdf[batchdf['date'].dt.year==y]
            temp = csvlist[i] 
            td = timedelta(j)
            #temp = temp[temp.index.isin(batch3['date']-td)]['AVG AIR TEMP oF']
            temp = (temp[temp.index.isin(batch3['date']-td)]['MAX AIR TEMP oF']-32)*5/9
            data = zip(temp.values,batch3['batches'].values)
            laglist.append(list(zip(temp.values,batch3['batches'].values)))
            #print(list(zip(ppt.values,batch['batches'].values)))
    laglist = sum(laglist,[])
    #print(laglist)
    lag = pd.DataFrame(laglist,columns = ['lag'+str(j), 'batch'])        
    lagdays.append(lag)
```

```python
for i in lagdays:
    i = i.set_index('batch',inplace=True)
lagdf = pd.concat(lagdays,axis=1,sort=False)
lagdf = lagdf.reset_index()
```

```python
############Find correlations for all lagdays###########
import numpy as np
corlist=[]
batch1temp=[] #mean of temp amounts with different lagdays when the 1 batch of mosquitoes is found
batch2_5temp=[]
batch6temp=[]
batch1meantemp=[] #Mean temp amount for each lagdays (1-14) when batch ==1

for i in range(1,len(lagdf.columns)):
    sordf =lagdf.sort_values(by='batch')    
    cor,_ = pearsonr(sordf[sordf.columns[i]],sordf['batch'])
    corlist.append(cor)
    mylist = [sordf[sordf['batch']==1][sordf.columns[j]].values for j in range(1,i+1)]
    mylist2 = [sordf[(sordf['batch']>1) & (sordf['batch']<6)][sordf.columns[j]].values for j in range(1,i+1)]
    mylist3 = [sordf[sordf['batch']>=6][sordf.columns[j]].values for j in range(1,i+1)]
    #batch1ppt.append(sordf[sordf['batch']==1][sordf.columns[i]].values)
    batch1temp.append(sum(map(np.array, mylist))/i)
    batch2_5temp.append(sum(map(np.array, mylist2))/i)
    batch6temp.append(sum(map(np.array, mylist3))/i)
    batch1meantemp.append(sordf[(sordf['batch']==1) & (sordf[sordf.columns[0]]!=0)][sordf.columns[0]].mean())
    #batch1ppt0.append(sordf[(sordf['batch']==1) & (sum(map(np.array, mylist))==0)]['batch'].sum())
def frqdistlagdays(day):
    plt.hist(batch1temp[day])
    plt.hist(batch2_5temp[day])
    plt.hist(batch6temp[day])
    #plt.title('lagday'+str(day))
    #plt.xlabel('mean temp.(oF) of'+" "+str(day)+' '+'lagdays')
    plt.title('Frequency distribution- '+str(day)+' day lagperiod', fontsize=16)
    plt.xlabel('mean max.temp.(oF)',fontsize=15)
    plt.ylabel('No. of mosquito batches',fontsize=15)
    plt.legend(['Batch=1','Batch=2-5','Batch>=6'])
```

```python
#figure(figsize=(15,20))
for i in range(1, 15):
    figure(figsize=(25,30))
    plt.subplot(5, 3, i)
    plt.ylim(0,30)
    plt.xlim(35,110)
    frqdistlagdays(i-1)
```
