```python
import pandas as pd
import matplotlib.pyplot as plt
blar03 = pd.DataFrame(buckdata(2003).sum(axis=1))
blar04 = pd.DataFrame(buckdata(2004).sum(axis=1))
blar05 = pd.DataFrame(buckdata(2005).sum(axis=1))
```

```python
blar = pd.concat([blar03,blar04,blar05])
b = b.set_index('date')
```

```python
import numpy as np
batchlar = blar.join(b).replace(np.nan,0).rename(columns={0:'larvae','batches':'WNV +ve'})
batchlar= batchlar.drop('datemd',axis=1)
```

```python
for i in range(2003,2006):
    ax = batchlar[batchlar.index.year==i].plot(secondary_y='WNV +ve',marker='o',ylim=(0,9),ylabel='WNV +ve adult mosquitoes')
    ax.set_ylim(0,1000)
    ax.set_ylabel('Larvae')
    plt.show()
```
