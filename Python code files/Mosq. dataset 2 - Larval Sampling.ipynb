```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import os, glob
os.chdir("/data/dhence/a/pborse2/inputdata/Mosquito data")

flist = [pd.read_csv(f) for f in glob.glob('*.csv')]
df1 = flist[5][flist[5]["Community_ID"]=='Springfield']

df1.index = range(1,11)
df1['sum']=df1[df1.columns[6:]].sum(axis=1)
df1['Date']= pd.to_datetime(df1['Date'])

ppt = pd.read_csv("/data/dhence/a/pborse2/inputdata/ISWS_data/Springfield/2012.csv")
datelist = pd.date_range(start = "2012-01-01", periods=len(ppt['DAY'])).tolist()
ppt['date']= datelist
ppt= ppt.set_index('date')

plt.scatter(df1['Date'],df1['sum'], marker = 'o')
plt.legend('mosq.')
from matplotlib.pyplot import figure
#figure(figsize=(15,6))
ax = ppt['TOTAL PRECIP IN'][120:270].plot(secondary_y=True, color='k')
ax.set_ylabel("Precipitation (in)")
plt.show()
```
