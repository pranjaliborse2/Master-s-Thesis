```python
import netCDF4
import xarray as xr
import numpy as np
import pandas as pd
import os, glob
import matplotlib.pyplot as plt
import geopandas as gpd
from datetime import timedelta
from scipy.stats import pearsonr, spearmanr
import cartopy.crs as ccrs
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
lar03 = pd.DataFrame(buckdata(2003).sum(axis=1)[buckdata(2003).sum(axis=1).index.month.isin(range(6,10))])
lar04 = pd.DataFrame(buckdata(2004).sum(axis=1)[buckdata(2004).sum(axis=1).index.month.isin(range(6,10))])
lar05 = pd.DataFrame(buckdata(2005).sum(axis=1)[buckdata(2005).sum(axis=1).index.month.isin(range(6,10))])
```

```python
lar03w = lar03.resample('W').sum()
lar04w = lar04.resample('W').sum()
lar05w = lar05.resample('W').sum()
```

```python
csv = [pd.read_csv(f) for f in glob.glob("/data/dhence/a/pborse2/inputdata/ISWS_data/Champaign/*.csv")]
from ipynb.fs.defs.ISWSdataset_Champaign import aprnov
csvlist = [aprnov(csv)[i] for i in range(len(csv))]
os.chdir('/data/dhence/a/pborse2/inputdata/data.eol.ucar.edu/pub/download/extra/katz_data/stage4/daily_data')
fi = [f for f in glob.glob('*.nc')]
fi.sort()
```

```python
temp = pd.concat(csvlist)[['MAX AIR TEMP oF']]
```

```python
temp0305 = temp[temp.index.year.isin(range(2003,2006))]
```

```python
tempw = temp0305.resample('7D').mean()
```

```python
rain = projection(fi)['tp'].sel(x=slice(734.5,737.5),y=slice(488,490.5)).mean(('x','y'))
missing = "2006-09-04T12:00:00.000000000"
miss1 = "2003-08-26T12:00:00.000000000"
# create an "empty" time slice with fillvalues
empty = xr.full_like(rain.isel(time=0), 0)
emp1 = xr.full_like(rain.isel(time=0), 0)
# fix the time coordinate of the timeslice
empty['time'] =pd.date_range(missing, periods=1)[0]
emp1['time']=pd.date_range(miss1, periods=1)[0]
# concat and sort time
rain = xr.concat([rain, empty,emp1], "time").sortby("time")
```

```python
rainw = rain[(rain.time.dt.year.isin(range(2003,2006)))&(rain.time.dt.month.isin(range(4,10)))].resample(time='7D').sum()
```

```python
lar=lar03w.append(lar04w.append(lar05w))
```

```python
rjj = [[rainw.sel(time=i-timedelta(weeks=j),method='nearest').values for j in range(11)] for i in lar[lar.index.month.isin([6,7])].index]
ras = [[rainw.sel(time=i-timedelta(weeks=j),method='nearest').values for j in range(11)] for i in lar[lar.index.month.isin([8,9])].index]
```

```python
tjj = [[tempw['MAX AIR TEMP oF'][tempw.index.get_loc(i-timedelta(weeks=j), method='nearest')] for j in range(11)] for i in lar[lar.index.month.isin([6,7])].index]
tas = [[tempw['MAX AIR TEMP oF'][tempw.index.get_loc(i-timedelta(weeks=j), method='nearest')] for j in range(11)] for i in lar[lar.index.month.isin([8,9])].index]
```

```python
from itertools import accumulate
import operator
rcumjj = [list(accumulate(i, operator.add)) for i in rjj]
rcumas = [list(accumulate(i, operator.add)) for i in ras]
tmeanjj = [list(accumulate(i,lambda x,y:(x+y)/2)) for i in tjj]
tmeanas = [list(accumulate(i,lambda x,y:(x+y)/2)) for i in tas]
```

```python
from scipy.stats import spearmanr
for k in [lar03w,lar04w,lar05w]:
    R = [spearmanr([tmeanjj[i][j] for i in range(len(k[k.index.month.isin([6,7])]))],k[k.index.month.isin([6,7])]['p'])[0] for j in range(11)]
    plt.plot(range(11),R)
```

```python
for larv in [lar03w,lar04w,lar05w]:
    r = [rain.sel(time= slice(i-timedelta(7),i-timedelta(1))).sum().values for i in larv.index]
    larv['cumrain7']=r
    t = [temp['MAX AIR TEMP oF'][temp.index.get_loc(i-timedelta(weeks=2), method='nearest')].mean() for i in larv.index]
    larv['meantemp7']=t
dflar = pd.concat([lar03w,lar04w,lar05w])
```

```python
from scipy.signal import find_peaks
peaklarv=[]
for larv in [lar03w,lar04w,lar05w]:
    peaks, _ = find_peaks(larv[larv.index.month.isin(range(6,10))][0], height=0)
    peaks1=larv[larv.index.month.isin(range(6,10))][0][peaks]
    notpeaks = larv[0][set(range(len(larv)))-set(peaks)]
    peaklarv.append(peaks1)
peaklarv1 = pd.concat(peaklarv)
dflar=dflar[dflar.index.month.isin(range(6,10))]
dflar['bool']='false'
for i in range(len(dflar)):
    if(list(peaklarv1.index.values).count(dflar.index[i])>0):
        dflar['bool'][i]='true'
```

```python
def prob(boolean,clim):
    for j in range(2003,2006):
        dflar2 = dflar[dflar.index.year==j]
        rainint = range(int(dflar2[clim].min()),int(dflar2[clim].max()),3)
        nrgrtint = [len(dflar2[dflar2[clim]>=i]) for i in rainint]
        nopeaks_rgrtint=[]
        for i in rainint:
            dfnopeaks = dflar2[dflar2[clim]>=i]
            l = len(dfnopeaks[dfnopeaks['bool']==boolean])
            nopeaks_rgrtint.append(l)
        print(nrgrtint,nopeaks_rgrtint)
        prob = [nopeaks_rgrtint[i]/nrgrtint[i] for i in range(len(nrgrtint)-1)]
        plt.plot(rainint[:-1],prob)
    plt.legend(range(3,6))
```

```python
## Rolling avg.
temproll = temp0305.rolling(21).mean()
rainroll = rain.rolling(time=1).sum()
larroll = (lar03.append(lar04.append(lar05)))#.rolling(7).sum()
```

```python
from datetime import datetime
for j in range(1,20): 
    larroll['meantemp'+str(j)]= temproll[temproll.index.isin(larroll.index-timedelta(j))].values
    larroll['cumrain'+str(j)] = rainroll[rainroll.time.dt.date.isin((larroll.index-timedelta(j)).date)]
```

```python
from scipy.signal import find_peaks
peaklarv=[]

peaks, _ = find_peaks(larroll[larroll.index.month.isin(range(6,10))][0], height=0)
peaks1=larroll[larroll.index.month.isin(range(6,10))][0][peaks]
notpeaks = larroll[0][set(range(len(larroll)))-set(peaks)]
peaklarv.append(peaks1)
peaklarv1 = pd.concat(peaklarv)

larroll['bool']='false'
for i in range(len(larroll)):
    if(list(peaklarv1.index.values).count(larroll.index[i])>0):
        larroll['bool'][i]='true'
```

```python
def prob(boolean,clim):
    for j in range(2003,2006):
        dflar2 = larroll[larroll.index.year==j]
#         rainint = [0,5,10,15,20,25,40,60,80]
        rainint = range(int(dflar2[clim].min()),int(dflar2[clim].max()),3)
        nrgrtint = [len(dflar2[dflar2[clim]>=i]) for i in rainint]
        nopeaks_rgrtint=[]
        for i in rainint:
            dfnopeaks = dflar2[dflar2[clim]>=i]
            l = len(dfnopeaks[dfnopeaks['bool']==boolean])
            nopeaks_rgrtint.append(l)
#         print(nrgrtint,nopeaks_rgrtint)
        prob = [nopeaks_rgrtint[i]/nrgrtint[i] for i in range(len(nrgrtint)-2)]
        plt.plot(rainint[:-2],prob)
    plt.legend(range(3,6)) 
```

```python
for i in range(1,20):
    prob('true','meantemp'+str(i))
    plt.show()
```
