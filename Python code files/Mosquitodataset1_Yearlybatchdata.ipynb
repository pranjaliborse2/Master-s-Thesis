```python
def yearlyfile(file,loc): # here loc = locations which are given as the sheetnames of csv
    from datetime import datetime
    os.chdir("/data/dhence/a/pborse2/inputdata/Mosquito data")
    mos = pd.read_excel(file, sheet_name=loc)
    mos['m-d']= [str(i.month)+'/'+str(i.day) for i in mos['Date Collected'].dt.date]
    mosyr = list(dict.fromkeys(mos['Date Collected'].dt.year))
    return mos, mosyr
```

```python
mos = yearlyfile("Mosquito yearly.xlsx",'U-C')[0]
mosyr = yearlyfile("Mosquito yearly.xlsx",'U-C')[1]
```

```python
mosyr
```

```python
#count total mosquito batches collected in a year
def batch1(year,mosfile):
    batch = [] #yearly count of mosquito batches
    for i in year:
        s=0
        for j in range(len(mosfile)):
            if(mosfile['Date Collected'][j].year==i):
                s=s+1
        batch.append(s)
    return batch
```

```python
batch = batch1(mosyr,mos)
batch
```

```python
def batchdf1(mosfile):
    dat=list(dict.fromkeys(mosfile['Date Collected']))
    datbatch = [] #datewise count of mosquito batches
    datmd=[]
    for i in dat:
        s=0
        for j in range(len(mosfile)):
            if(mosfile['Date Collected'][j]==i):
                s=s+1
                md = mosfile['m-d'][j] 
        datbatch.append(s)
        datmd.append(md)
    batchdf = pd.DataFrame({'batches':datbatch,'date':dat, 'datemd':datmd})
    return batchdf
```

```python
batchdf = batchdf1(mos)
```

```python
#Average consequtive days of precipitation in a year - Observe the trend over the years
condayyr =[] # Number of consecutive rain days in each year
totppt =[] # This list contains total rain days in a year year
for df in csvlist:
    s=0
    conday=[] #No. of Consecutive rain days in individual year
    t=0 #parameter to calculate total pptdays in a year
    for i in range(0,len(df['TOTAL PRECIP oF']),31):
        con=[] #No. of Consecutive rain days in individual month
        for ppt in df['TOTAL PRECIP oF'][i:i+31]:
            if(ppt>0):
                s=s+1
                t=t+1
            else:
                if(s>1):
                    con.append(s)
                s=0
        conday.append(con)
    condayyr.append(conday)
    totppt.append(t)
```
